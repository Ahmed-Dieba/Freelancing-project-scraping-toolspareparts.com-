```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
lis = lis = ['https://www.toolspareparts.com.au/aeg-abg5520-spare-parts.html',
'https://www.toolspareparts.com.au/aeg-jsr18240-spare-parts.html',
'https://www.toolspareparts.com.au/aeg-br18c-x4-spare-parts.html',
'https://www.toolspareparts.com.au/aeg-br18c-spare-parts.html',
'https://www.toolspareparts.com.au/aeg-psu1000-spare-parts.html',
'https://www.toolspareparts.com.au/aeg-ps305dg-spare-parts.html',
'https://www.toolspareparts.com.au/aeg-ps254sb-spare-parts.html',
'https://www.toolspareparts.com.au/aeg-msc305-spare-parts.html',
'https://www.toolspareparts.com.au/aeg-eac1220w-spare-parts.html',
'https://www.toolspareparts.com.au/aeg-sfnc1965-spare-parts.html',
'https://www.toolspareparts.com.au/aeg-chfn90-spare-parts.html',
'https://www.toolspareparts.com.au/aeg-cfn90-spare-parts.html',
'https://www.toolspareparts.com.au/aeg-afnda2565-spare-parts.html',
'https://www.toolspareparts.com.au/aeg-afnda18-spare-parts.html',
'https://www.toolspareparts.com.au/aeg-agn6500-spare-parts.html',
'https://www.toolspareparts.com.au/aeg-agn6000db-spare-parts.html',
'https://www.toolspareparts.com.au/aeg-agn5000deb-rev2-spare-parts.html',
'https://www.toolspareparts.com.au/aeg-agn5000deb-spare-parts.html',
'https://www.toolspareparts.com.au/aeg-agn3800b-spare-parts.html']
Url_list = []
Name_list = []
sku_list = []
image_url_list= []
Image_name_list= []
diagram_url_list= []
diagram_name_list= []
Details_list= []
Diagram_list= []
Diagram_Product_No_list= []
Product_Name_list= []
Price_list= []

df = {
    'URL': Url_list,
    'Name':Name_list,
    'sku':sku_list,
    'image url':image_url_list,
    'Image name':Image_name_list,
    'diagram url':diagram_url_list,
    'diagram name':diagram_name_list,
    'Details':Details_list,
    'Diagram':Diagram_list,
    'Diagram Product No' :Diagram_Product_No_list,
    'Product Name':Product_Name_list,
    'Price':Price_list
}

for url in lis:
    source = requests.get(url).text
    soup = BeautifulSoup(source,'lxml')
    Url_list.append(url)
    Name = soup.find_all('span', class_ = 'base')[0].text
    Name_list.append(Name)
    sku = soup.find('div', class_='value').string
    sku_list.append(sku)
    image = soup.find('img', class_='gallery-placeholder__image').get('src')
    image_url_list.append(image)
    Image_name = image.split('/')[-1]
    Image_name_list.append(Image_name)
    diagram_url = soup.find('a',class_='diagram__link')
    if diagram_url == None:
        diagram_url = 'Na'
    else:
        diagram_url = diagram_url.get('href')
    diagram_url_list.append(diagram_url)
    diagram_name = diagram_url.split('/')[-1]
    diagram_name_list.append(diagram_name)
    Details = Name
    Details_list.append(Name)
    rows = soup.find('tbody')
    rows = rows.find_all('tr')
    
    Diagram = ''
    for row in rows:
        Diagram += row.find_all('td')[0].string + "\n"
    Diagram_list.append(Diagram)
    
    Diagram_Product_No = ''
    for row in rows:
        temp = row.find_all('td')
        if len(temp) ==1:
            Diagram_Product_No += "Na"+ "\n"
        elif temp[1].string == None :
          Diagram_Product_No += "Na"+ "\n"
        else:
          Diagram_Product_No += temp[1].string + "\n"     
    Diagram_Product_No_list.append(Diagram_Product_No)
    
    Product_Name = ''
    for row in rows:
        temp = row.find('strong',class_='product-item-name')
        if temp == None:
            Product_Name += "Na"+ "\n"
        else:
            Product_Name += temp.string +'\n'
            
    Product_Name_list.append(Product_Name)
    
    Price = ''
    for row in rows:
        temp = str(row.find('span',class_='price'))
        temp = temp.split('$')[-1].split('<')[0]
        Price += temp + '\n'
    Price_list.append(Price)
```
